```python
# this notebook demonstrates how to use SparkSQL on a dataframe
# 
```

```python
from pyspark import SparkConf, SparkContext
from pyspark.sql import SQLContext
```

```python
from pyspark.sql.session import SparkSession
sc = SparkContext('local')
spark = SparkSession(sc)
```

```python
sqlContext = SQLContext(sc)
```

```python
df = sqlContext.read.format("csv").option("inferschema","true").option("header", "true").option("delimiter", ",").load("gapminder_all_binary.csv")
```

```python
# Register the DataFrame as a SQL temporary view
# see 
df.createOrReplaceTempView("df")
```

```python
query = '''
SELECT * from df
'''
```

```python
out = spark.sql(query)
```

```python
out.columns
```

```python
out.show()
```

```python
query = '''
SELECT continent, AVG(gdpPercap_2007) FROM df GROUP BY continent
'''
```

```python
avg_gdp = spark.sql(query)
```

```python
avg_gdp.show()
```
